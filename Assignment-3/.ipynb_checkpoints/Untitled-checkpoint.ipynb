```python
M_S = 0.15
M_C = 0.001
```

```python
from random import random
x = open("ratings.csv")
lst = [[]]*615
totsize = 0
for id, row in enumerate(x.readlines()):
    if id==0:
        continue
    row = row.split(',')
    if float(row[2])<2.0:
        continue
    lst[int(row[0])] = lst[int(row[0])] + [int(row[1])]
fin = []
train = []
test = []
cnt = dict()
for ele in lst:
    if len(ele)<10:
        continue
    totsize+=1
    trn = []
    tst = []
    for i in ele:
        if random()<0.2:
            tst.append(i)
        else:
            if i in cnt:
                cnt[i]+=1
            else:
                cnt[i] = 1
            trn.append(i)
    train.append(trn)
    test.append(tst)
```

```python
L = dict()
for x, y in cnt.items():
    if(float(y)/float(totsize)<M_S):
       continue
    L[(x,)] = y
LL = []
LL.append(L)
```

```python
from itertools import combinations
def has_infrequent_itemset(c, L):
    for s in combinations(c, len(c) - 1):
        if s not in LL[-1]:
            return True
    return False

def apriori_gen(L):
    C = []
    for x, y in combinations(L, 2):
        if not x[:-1] == y[:-1]:
            continue
        nxt = tuple(sorted(list(x) + [y[-1]]))
        if has_infrequent_itemset(nxt, L):
            continue
        C.append(nxt)
    return C
```

```python
from itertools import combinations
for k in range(2, 10):
    if len(LL[-1]) == 0:
        print(f"Highest frequent itemset size: {k - 1}")
        break
        
    Ck = apriori_gen(LL[- 1])
    if(len(Ck)>0):
        print(Ck[0])
    new_L = dict()
    for c in Ck:
        curr = 0
        for tr in train:
            curr += set(c).issubset(tr)
        if(float(curr)/float(totsize)<M_S):
            continue
        new_L[c] = curr
    
    LL.append(new_L)
```

```python
rules = [] # (x, y, support, confidence)
for k in range(1, len(LL)):
    for xy, cnt in LL[k].items():
        for x in xy:
            y = set(xy)
            y.remove(x)
            x, y = tuple([x]), tuple(y)
            rules.append((x, y, cnt/totsize, cnt/LL[0][x]))
rules.sort(reverse = True, key = lambda t: t[2])
top_support = rules[:100]
rules.sort(reverse = True, key = lambda t: t[3])
top_confidence = rules[:100]
common_rules = []
for x in top_confidence:
    if x in top_support:
        common_rules.append(x)
```

```python
from collections import defaultdict
def predict(movieIds, test_predictions, N):
    useful_rules = []
    for x, y, supp, conf in rules:
        if x[0] not in movieIds:
            continue
        new_y = [movieId for movieId in y if movieId not in movieIds]
        if len(new_y) == 0:
            continue
        if (x[0], new_y) not in useful_rules:
            useful_rules.append((x[0], new_y))
        if len(useful_rules) == N:
            break
    
    predictions = set()
    precision = 0
    recall = 0
    for n, (x, y) in enumerate(useful_rules):
        predictions = predictions.union(set(y))
        hit_set = predictions & test_predictions
        precision = len(hit_set) / len(predictions)
        recall = len(hit_set) / len(test_predictions)
#         print(f"Taking {n} rules\t precision: {precision:.3f}\t recall: {recall:.3f}")
    
    return precision, recall
recommend = defaultdict(list)
N = 10
average_precision = [0 for _ in range(N)]
average_recall = [0 for _ in range(N)]
for index, row in enumerate(train):
    # print(f"{'-'*100}\nUserId: {row['userId']}")
    test_predictions = test[index]
    for i in range(N):
        if(len(test_predictions)==0):
            continue
        prec, recall = predict(row, set(test_predictions), i + 1)
        # print(prec, recall)
        average_precision[i] += prec
        average_recall[i] += recall
        
average_precision = map(lambda x: x / totsize, average_precision)
average_recall = map(lambda x: x / totsize, average_recall)
import pandas as pd
df = pd.DataFrame(list(zip(average_precision, average_recall)), columns=["precision", "recall"])
import plotly.express as px
fig = px.line(df, x=df.index, y="precision")
fig.show()
fig2 = px.line(df, x=df.index, y="recall")
fig2.show()
```
