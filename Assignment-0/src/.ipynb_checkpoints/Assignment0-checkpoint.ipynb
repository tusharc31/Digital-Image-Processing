Yes, the images seem unusual and and there seems to be some sort of inversion compared to the original images. 

The issue is that OpenCV uses BGR as its default colour order for images whereas matplotlib uses RGB. So when we display an image loaded with OpenCV using matplotlib, the order of the channels is incorrect. To fix the issue, we explicitly convert our OpenCV image to RGB.

```python
image1 = cv2.imread('../images/Image1.jpg')
image2 = cv2.imread('../images/Image2.jpg')
image3 = cv2.imread('../images/Image3.jpg')
image4 = cv2.imread('../images/Image4.jpg')

image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
image3 = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)
image4 = cv2.cvtColor(image4, cv2.COLOR_BGR2RGB)

plt.subplot(2, 2, 1)
plt.imshow(image1)

plt.subplot(2, 2, 2)
plt.imshow(image2)

plt.subplot(2, 2, 3)
plt.imshow(image3)

plt.subplot(2, 2, 4)
plt.imshow(image4)

plt.show()
```

### Question

Convert any color image (of your choice) to grayscale without using a function and save
that image.

**Method adopted** - To get the grey-scale image, I've used the Luminosity Method, which takes a weighted average of the RGB channel. The weights for the colors are different because we know that our eyes react to each color in a different manner. Specifically, our eyes are more sensitive to green, then to red, and finally to blue. Hence, the weights used are 0.59 for green, 0.3 for red and 0.11 for blue respectively.

```python
image1 = cv2.imread('../images/Image1.jpg')
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
plt.imshow(image1)
plt.title("Original Image")
plt.show()

blue_channel  = image1[:,:,0]
green_channel = image1[:,:,1]
red_channel   = image1[:,:,2]

grey_img = 0.3*red_channel + 0.59*green_channel + 0.11*blue_channel
grey_img = np.stack((grey_img,)*3, axis=-1)
grey_img = grey_img.astype(int)
plt.imshow(grey_img)
plt.title("Black and White Image")
cv2.imwrite("../results/black_ans_white.jpg", grey_img)
plt.show()
```

### Question

Check the range of values in the image. What do you infer from this? Include your answer in a markdown cell.

```python
red_channel   = red_channel.flatten()
blue_channel  = blue_channel.flatten()
green_channel = green_channel.flatten()

def analyze_channel(channel, color):
    print("Minimum Intensity: ", np.min(channel))
    print("Maximum Intensity: ", np.max(channel))
    print("Mean Intensity: ", np.mean(channel))
    print("Standard deviation: ", np.std(channel))
    _ = plt.hist(channel, color = color, bins='auto')
    plt.title("Histogram for intensity distribution")
    plt.xlabel("Intensity")
    plt.ylabel("Number of pixels")
    plt.show()
```

#### Analysing the red channel

```python
analyze_channel(red_channel, "red")
```

#### Analysing the green channel

```python
analyze_channel(green_channel, "green")
```

#### Analysing the blue channel

```python
analyze_channel(blue_channel, "blue")
```

**Observations** - We can notice that the original image mainly comprises of brown and blue sections. Since, brown is a mixture of red and green, the statistics for red and green are almost similar, and blue seems independent. The average intensity in all channels are approximately in the range of 70 to 90, which means that the image is slightly on the darker side. However, the min and max values in all channels are 0 and 255 respectively, which I think arise due to presence of black and white colored pixels respectively. 

#### Analysing the complete image

```python
analyze_channel(image1.flatten(), "grey")
```
